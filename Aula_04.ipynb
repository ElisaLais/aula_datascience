## Desafio 3 do [Paulo Silveira](https://twitter.com/paulo_caelum)


Pairplot dos acertos de cada categoria (CN, CH, MT, LC, nota pura da redação). Usar o gabarito e as respostas

```
def conta_respostas (resposta, gabarito):
  if (pd.isnull(resposta)) | (pd.isnull(gabarito)):
    return -1
  count = 0
  erro = 0
  for i in range(len(resposta)):
    if resposta[i] == gabarito[i]:
      count = count + 1
  return count
```

```
#respostas_mt = dados['TX_RESPOSTAS_MT'].dropna().str.replace(r"([A-Z])", r" \1").str.split(expand=True)
#gabarito_mt = dados['TX_GABARITO_MT'].dropna().str.replace(r"([A-Z])", r" \1").str.split(expand=True)
#dados = dados.dropna(subset=['TX_RESPOSTAS_MT','TX_GABARITO_MT'])


dados['Acertos_MT'] = dados.apply(lambda aluno : conta_respostas(aluno['TX_RESPOSTAS_MT'],aluno['TX_GABARITO_MT']),axis=1)
dados['Acertos_CN'] = dados.apply(lambda aluno : conta_respostas(aluno['TX_RESPOSTAS_CN'],aluno['TX_GABARITO_CN']),axis=1)
dados['Acertos_CH'] = dados.apply(lambda aluno : conta_respostas(aluno['TX_RESPOSTAS_CH'],aluno['TX_GABARITO_CH']),axis=1)
dados['Acertos_LC'] = dados.apply(lambda aluno : conta_respostas(aluno['TX_RESPOSTAS_LC'],aluno['TX_GABARITO_LC']),axis=1)
# Para cada aluno irá pegar uma resposta e aplicar a função conta resposta
```

```
dados_acertos = dados[['Acertos_MT','Acertos_CN','Acertos_CH','Acertos_LC','NU_NOTA_REDACAO']]
dados_acertos = dados_acertos.dropna()
```

```
sns.pairplot(dados_acertos) 
```

#Observações
É possível observar 3 tendêndicas interessantes. Ao mesmo tempo que o número de acertos em uma determinada matéria como Matemática (MT) não determina seu número de acertos em Ciências da Natureza (CN), porém observa-se que os alunos com maior número de acertos na amostra (aprox 45 pts) seguiram este padrão em todas as provas do ENEM exceto na prova de Redação. Por fim, podemos observar com a última linha de gráficos que o número de acertos nas diferentes provas possui baixa correlação com a nota de redação do candidato. 

```
corr_acertos = dados_acertos.corr()

matrix = np.triu(corr_acertos)
sns.heatmap(corr_acertos, annot=True, mask=matrix, cmap='RdBu', square=True)
```

## Desafio 4 do [Guilherme Silveira](https://twitter.com/guilhermecaelum)

Remover todos os zeros. Tomar o cuidado que no desafio 1 já tomamos decisões ligadas a limpeza dos dados também. Você também pode exportar para outro CSV se quiser.

```
dados_notas.head()
```

```
#Esta função troca a nota zero pela média da categoria
def trocar_zeros(nota,media):
  for i in range(len(dados_notas)):
    if nota == 0:
      nota = media
    return nota
```

```
#Extraindo a media de cada categoria
med_CN = dados_notas['ciencias_naturais'].mean()
med_CH = dados_notas['ciencias_humanas'].mean()
med_LC = dados_notas['linguagem_codigo'].mean()
med_MT = dados_notas['matematica'].mean()
med_redacao = dados_notas['redacao'].mean()
```

```
#Troca-se a nota zero do aluno para a média geral da prova

dados_notas['ciencias_naturais'] = dados_notas.apply(lambda aluno : trocar_zeros(aluno['ciencias_naturais'],med_CN),axis=1)
dados_notas['ciencias_humanas'] = dados_notas.apply(lambda aluno : trocar_zeros(aluno['ciencias_humanas'],med_CH),axis=1)
dados_notas['linguagem_codigo'] = dados_notas.apply(lambda aluno : trocar_zeros(aluno['linguagem_codigo'],med_LC),axis=1)
dados_notas['matematica'] = dados_notas.apply(lambda aluno : trocar_zeros(aluno['matematica'],med_MT),axis=1)
dados_notas['redacao'] = dados_notas.apply(lambda aluno : trocar_zeros(aluno['redacao'],med_redacao),axis=1)
```

```
dados_notas.head()
```

```
sns.pairplot(dados_notas)
```

```
dados_notas.describe()
```

```
corr_notas = dados_notas.corr()

matrix = np.triu(corr_notas)
sns.heatmap(corr_notas, annot=True, mask=matrix, cmap='RdBu', square=True)
```

#Observação
Foram consideradas apenas os valores que já encontravam-se na base como forma de simular o estado real do fechamento das notas do exame. Como foi possível observar, zerar uma prova não desclassifica o estudante do exame, neste sentido, para corrigir os zeros adotei a abordagem de substituição pela média geral de cada prova. Como é possível observar pela matriz de correlação, a retirada dos valores nulos não resultou em uma melhora. 

É possível dizer que a Teoria de Resposta ao Item (TRI) esteja gerando este efeito, pois existem diferentes pontuações para diferentes tipos de questões.
